```python
# Importar librerías 

import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import numpy as np

%matplotlib inline

matplotlib.rcParams['figure.figsize'] = (12,8) # Ajustar las configuraciones en las gráficas creadas

# Leer la información

df = pd.read_csv('/Users/joseabjerez/Downloads/movies.csv')
```

```python
# Revisar la info en df

df.head()
```

```python
# Encontrar espacios en blanco o vacíos 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Tipos de datos en las columnas

df.dtypes
```

```python
# Rellenar los valores NA en columna budget, runtime, votes y gross con un cero
df['budget'].fillna(0, inplace=True)
df['gross'].fillna(0, inplace=True)
df['votes'].fillna(0, inplace=True)
df['runtime'].fillna(0, inplace=True)
```

```python
# Cambiar tipo de datos en columna budget, runtime, votes y gross
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
df['votes']= df['votes'].astype('int64')
df['runtime']= df['runtime'].astype('int64')
```

```python
df.corr(numeric_only= True)

corr_matrix= df.corr(numeric_only= True)

sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
