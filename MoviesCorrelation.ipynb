```python
# Importar librerías
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import numpy as np

# Configuración para que las gráficas se muestren directamente en el notebook
%matplotlib inline

# Configuración del tamaño predeterminado de las figuras
matplotlib.rcParams['figure.figsize'] = (12, 8)

# Leer la información desde un archivo CSV. Asegúrate de que la ruta sea correcta.
df = pd.read_csv('/Users/joseabjerez/Downloads/movies.csv')
```

```python
# Iterar sobre las columnas del DataFrame
for col in df.columns:
    # Calcular el porcentaje de valores nulos o vacíos en la columna actual
    pct_missing = np.mean(df[col].isnull())
    
    # Imprimir el nombre de la columna y el porcentaje de valores nulos
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Rellenar los valores NA en la columna 'budget' con cero
df['budget'].fillna(0, inplace=True)

# Rellenar los valores NA en la columna 'gross' con cero
df['gross'].fillna(0, inplace=True)

# Rellenar los valores NA en la columna 'votes' con cero
df['votes'].fillna(0, inplace=True)

# Rellenar los valores NA en la columna 'runtime' con cero
df['runtime'].fillna(0, inplace=True)
```

```python
# Cambiar el tipo de datos de la columna 'budget' a int64
df['budget'] = df['budget'].astype('int64')

# Cambiar el tipo de datos de la columna 'gross' a int64
df['gross'] = df['gross'].astype('int64')

# Cambiar el tipo de datos de la columna 'votes' a int64
df['votes'] = df['votes'].astype('int64')

# Cambiar el tipo de datos de la columna 'runtime' a int64
df['runtime'] = df['runtime'].astype('int64')
```

```python
# Ordenar el DataFrame por la columna 'gross' de manera descendente (mayor a menor)
# No se realiza la modificación en el DataFrame original (inplace=False)
sorted_df = df.sort_values(by=['gross'], inplace=False, ascending=False)

# Mostrar las primeras filas del DataFrame ordenado
top_movies = sorted_df.head()

# Imprimir el DataFrame resultante
print(top_movies)
```

```python
# Calcular la matriz de correlación para las características numéricas del DataFrame
corr_matrix = df.corr(numeric_only=True)

# Trazar un mapa de calor de la matriz de correlación con anotaciones
sns.heatmap(corr_matrix, annot=True)

# Configurar el título del mapa de calor
plt.title('Correlation Matrix for Numeric Features')

# Configurar las etiquetas de los ejes x e y
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

# Mostrar el mapa de calor
plt.show()
```

```python
# Aplicar la función factorize a cada columna y calcular la correlación de Pearson
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

# Trazar un mapa de calor de la matriz de correlación con anotaciones
sns.heatmap(correlation_matrix, annot=True)

# Configurar el título del mapa de calor
plt.title("Correlation matrix for Movies")

# Configurar las etiquetas de los ejes x e y
plt.xlabel("Movie features")
plt.ylabel("Movie features")

# Mostrar el mapa de calor
plt.show()
```

```python
# Aplicar la función factorize a cada columna y calcular la correlación
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

# Desapilar la matriz de correlación para obtener pares de correlación
corr_pairs = correlation_mat.unstack()

# Imprimir los primeros elementos de la serie de pares de correlación
print(corr_pairs.head())
```

```python
# Ordenar los pares de correlación en orden ascendente utilizando el método de ordenación "quicksort"
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

# Imprimir los pares de correlación ordenados
print(sorted_pairs)
```

```python
# Seleccionar pares de correlación cuyo valor absoluto sea mayor que 0.5
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

# Imprimir los pares de correlación fuertes
print(strong_pairs)
```
